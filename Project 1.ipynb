### Import libraries

## 1.Find all records where the weather was exactly clear.

```python
# df["Weather"].unique()
clear_weather = df[df["Weather"] == "Clear"]

clear_weather
```

## 2.Find the number of times the wind speed was exactly 4 km/hr.

```python
_4_wind_speed = df[df["Wind Speed_km/h"]==4].count()
print(_4_wind_speed.shape[0])
```

### 3.Check if there are any NULL values present in the dataset.

```python
null_counts = df.isnull().sum()
print(null_counts)
```

## 4.Rename the column "Weather" to "Weather_Condition."

```python
df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
print(df.head())
```

### 5.What is the mean visibility of the dataset?

```python
mean_vis = df["Visibility_km"].mean()
print(mean_vis)
```

### 6.Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.

```python
# df[df["Wind Speed_km/h"] > 24 ]

# df[df["Visibility_km"] == 25]

w_24_v_25 = df[(df["Wind Speed_km/h"] > 24) & (df["Visibility_km"] == 25)]

print(w_24_v_25.head())
```

### 7.What is the mean value of each column for each weather condition?

```python

mean_values = df.groupby('Weather_Condition').mean(numeric_only=True)
mean_values 
```

### 8.Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

```python
humid_visc =df[(df["Rel Hum_%"] > 50) | (df["Visibility_km"] > 40)]
# humid_visc.sample(n=10)
print(humid_visc.head())
```

### 9.Find the number of weather conditions that include snow.

```python
snow_conditions = df[df['Weather_Condition'].str.contains('Snow')]

# Count the unique snow conditions
num_snow_conditions = snow_conditions['Weather_Condition'].nunique()

print(num_snow_conditions)
```

### 1.Find all records where the weather was exactly clear.

```python
query ="SELECT * FROM weather_data WHERE Weather_Condition = 'Clear' LIMIT 5;"  
sql_df = pd.read_sql(query, engine)
print(sql_df)
```

### 2. Find the number of times the wind speed was exactly 4 km/hr.

```python
query ="SELECT * FROM weather_data WHERE `Wind Speed_km/h` =4 LIMIT 5;"  
sql_df = pd.read_sql(query, engine)
print(sql_df)
```

### 3.What is the mean visibility of the dataset?

```python
query ="SELECT AVG(Visibility_km) FROM weather_data;"  
sql_df = pd.read_sql(query, engine)
print(sql_df)
```

### 4.Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km

```python
query ="SELECT * FROM weather_data WHERE `Wind Speed_km/h` > 24 AND Visibility_km =25 LIMIT 5 ;"  
sql_df = pd.read_sql(query, engine)
print(sql_df)
```
